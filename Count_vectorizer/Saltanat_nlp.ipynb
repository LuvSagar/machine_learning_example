## What is Natural Language processing NLP ?
NLP in machine learning has many branches, suppose you wanted to predict the next word of a sentence or classify a book a review as negative or positive. These examples involve the data being in the form of words, depending on your task at hand you will need to represent your text data in a usable form so an algorithm can work.  

## What kind of data do we have at Yapily  and what part of NLP are we looking at?

At Yapily we look at transactional data, as you may know, this data is relatively short for each transaction and doesn't have any emotive meaning to it like a movie review (trying to class a reviews as negative or positive is a branch of NLP called Sentimental analysis). However, each transaction has a category (EATING_OUT, INCOME .. etc) and so at Yapily we are trying to build a model that will place an unknow transaction into the right category. For this to work we need to train a model on all the labelled data; so when that model sees a new transaction that has not been labelled we can place it in the right category at least with a high probability. For example, if the model sees TESCO then it will hopefully place that transaction in the Groceries category.

# What is the count Vectorizer?

The count vectorizer will count the amount of times a word has appeared in that paritcular document, note in our case document is transaction.

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
```

```python
cvec=CountVectorizer()
```

Note here the count vectorizer is a class here and i have assigned it to the variable cvec. 


```python
cvec.fit([transaction_1,transaction_2])
```

Note above that .fit() is a functions that belongs to the CountVectorizer class, For more on class and functions using python https://www.programiz.com/python-programming/class

```python
df = pd.DataFrame(cvec.transform([transaction_1,transaction_2]).toarray(),
                  columns=cvec.get_feature_names())


```

Only after fitting my CountVectorizer() over  can I transform the data, also note that cvec.transform([transaction_1,transaction_2]) will give a sparse matrix the toarray() function will convert that to normal 2 dimensional matrix which is put into a pandas dataframe df. For more on fit(),transform() and fit_trasform() functions check out https://towardsdatascience.com/fit-vs-transform-in-scikit-libraries-for-machine-learning-3c70e6300ded and for more on 
sparse matrix check out https://machinelearningmastery.com/sparse-matrices-for-machine-learning/

```python
df
```

Note the above pandas data frame we can see that there are only two rows, the first row represent transaction_1 and the second row represents transaction_2. The columns are all the different words found in both the transactions, as it can be seen direct and sports is only found once in the first transaction and tesco is not found any times in the first transaction. Note in the second row tesco is found only once and direct and sports are not present in that transaction.

Note Saltanat that you have labelled some transactions as "NAN" and this is not a category so we will have to drop those transactions from the data set. First, we have to convert those values to missing value so that makes it easier to drop.


```python
df.labels.value_counts()
```

The above code will replace "NAN" in the labels column with np.nan missing value, check now below what happens when you run value_counts. Note that np comes from the numpy library that was imported earlier.

```python
df.labels.value_counts()

```

Thats right the "NAN" labels are considered missing so pandas just doesn't count them. Now let's drop all the missing values in the data set before we Countvectorize it.

```python
print(df.shape)
```
